```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from ast import literal_eval
from collections import Counter
```

```python
#import processed data
lddf = pd.read_csv('new2_lddf.csv', converters = {
    'description': literal_eval, 'host_about': literal_eval, 
    'bigram_description': literal_eval, 'bigram_host_about': literal_eval})
nydf = pd.read_csv('new2_nydf.csv', converters = {
    'description': literal_eval, 'host_about': literal_eval, 
    'bigram_description': literal_eval, 'bigram_host_about': literal_eval})
tkdf = pd.read_csv('new2_tkdf.csv', converters = {
    'description': literal_eval, 'host_about': literal_eval, 
    'bigram_description': literal_eval, 'bigram_host_about': literal_eval})
```

## 2h. Bigram Analysis
<a id='Bigram'></a>

```python
lowld=lddf[lddf.review_scores_rating <=85]
highld=lddf[lddf.review_scores_rating == 100]
```

```python
print('London low rating group shape',lowld.shape)
print('London high rating group shape',highld.shape)
```

```python
lowny=nydf[nydf.review_scores_rating <=85]
highny=nydf[nydf.review_scores_rating ==100]
```

```python
print('New York low rating group shape',lowny.shape)
print('New York high rating group shape',highny.shape)
```

```python
lowtk=tkdf[tkdf.review_scores_rating <=85]
hightk=tkdf[tkdf.review_scores_rating ==100]
```

```python
print('Tokyo low rating group shape',lowtk.shape)
print('Tokyo high rating group shape',hightk.shape)
```

```python
#London description
ldcnt_low_description = Counter()
for item in lowld.bigram_description:
    for bigram in item:
        ldcnt_low_description[bigram]+=1
        
ldcnt_high_description = Counter()
for item in highld.bigram_description:
    for bigram in item:
        ldcnt_high_description[bigram]+=1
```

```python
ldlow_description_df = pd.DataFrame(ldcnt_low_description.most_common(10))
ldlow_description_df.columns = ['Bigram', 'Count']

ldhigh_description_df = pd.DataFrame(ldcnt_high_description.most_common(10))
ldhigh_description_df.columns = ['Bigram', 'Count']

fig, axes = plt.subplots(1,2, figsize = (14,6))
p = sns.barplot(x = 'Bigram', y = 'Count', data = ldlow_description_df, ax = axes[0])
p.set_xticklabels(labels = ldlow_description_df.Bigram, rotation=90)
p.set_title('Low Rating')

p = sns.barplot(x = 'Bigram', y = 'Count', data = ldhigh_description_df, ax = axes[1])
p.set_xticklabels(labels = ldhigh_description_df.Bigram, rotation=90)
p.set_title('High Rating')
```
